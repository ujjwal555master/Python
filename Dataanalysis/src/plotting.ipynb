# Color , marker and Line Style

```python
plt.plot(np.random.standard_normal(30).cumsum(), color="red", linestyle="dashed", marker="o")
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data1 = np.random.standard_normal(30).cumsum()
data2 = np.random.standard_normal(30).cumsum()
ax.plot(data1, color="black", linestyle="dotted", label="Default");
ax.plot(data2, color="blue", linestyle="dashed", label="steps-posts");
ax.legend()
```

```python
from datetime import datetime
```

```python
fig, ax = plt.subplots();
data = pd.read_csv("spx.csv")
data.head()
```

```python
data = pd.read_csv("spx.csv", index_col=0, parse_dates=True)
data.head()
```

```python
spx = data["SPX"]
```

```python
spx.plot(ax=ax, color="black")
fig
```

```python
crisis_data = [
    (datetime(2007, 10, 11), "peak of bull marke"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]
```

```python
crisis_data
```

```python
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+75), 
                xytext=(date, spx.asof(date)+225), 
                arrowprops=dict(facecolor="red", headwidth=4, width=2, headlength=4),
                horizontalalignment = "left", verticalalignment="top")
    
```

```python
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_title("Important dates in the 2008-2009 financial crisis")
```

```python
fig
```

# Shape plotting

```python
fig, ax = plt.subplots()
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="red", alpha=0.3)
ax.add_patch(rect)
ax.add_patch(circ)
```
